# 과제 : IQR 기반 & Z-score 기반 이상치 탐지
## AirQualityUCI_refined["PT08.S3(NOx)"]
### - Set Data

```python
import pandas as pd
import numpy as np
from pandas import datetime
from matplotlib import pyplot as plt
```

```python
# Load DataFile
def parser(x):
    return datetime.strptime(x, '%Y-%m-%d %H:%M:%S')

air = pd.read_csv('./data/AirQualityUCI_refined.csv',
                 index_col=[0],
                 parse_dates=[0],
                 date_parser=parser)
air.head()
```

```python
# PT08.S3(NOx)
pt = air["PT08.S3(NOx)"].copy()
print(pt.head())
pt.plot(figsize=(10,5))
```

```python
# Missing data - Linear Interpolation
pt.interpolate(inplace=True)

plt.figure(figsize=(10,5))
plt.plot(air["PT08.S3(NOx)"], zorder=2)
plt.plot(pt, zorder=1)
```

#
### - IQR-baesd

```python
# Boxplot
plt.boxplot(pt)
plt.xlabel("PT08.S3(NOx)")
```

```python
# Find IQR and fences
q1 = pt.quantile(0.25)
q2 = pt.median()
q3 = pt.quantile(0.75)

iqr = q3-q1
upper_fence = q3 + 1.5*iqr
lower_fence = q1 - 1.5*iqr

print(q1 ,q2, q3, iqr)
print(upper_fence, lower_fence)
```

```python
# Outlier detection
outliers_iqr = pt.loc[(pt>upper_fence)|(pt<lower_fence)]
outliers_iqr
```

```python
# Masking outliers
mask_iqr = pt.index.isin(outliers_iqr.index)
mask_iqr[:20]
```

```python
# Remove outliers
pt_refined_iqr = pt.copy()
pt_refined_iqr[mask_iqr] = np.nan
pt_refined_iqr
```

```python
# Linear interpolation(feat. outliers removed)
pt_refined_iqr.interpolate(inplace=True)
pt_refined_iqr.plot(figsize=(10,5))
```

#
### - Z-score

```python
import seaborn as sns

plt.figure(figsize=(10,5))
sns.distplot(pt)
# 정규분포처럼 보이지만 왼쪽으로 치우쳐 있음. right-skewed data.
```

```python
# Mean, Standard deviation
mean = np.mean(pt)
std = np.std(pt)
print(mean, std)
```

```python
# Calculate Z-scores
thres = 3

outliers_Z = pt.loc[np.abs((pt-mean)/std) > thres]
outliers_Z
```

```python
# Masking outliers
mask_Z = pt.index.isin(outliers_Z.index)
mask_Z[:20]
```

```python
# Remove outliers
pt_refined_Z = pt.copy()
pt_refined_Z[mask_Z] = np.nan
pt_refined_Z[mask_Z]
```

```python
# Linear interpolation(feat. outlier removed)
pt_refined_Z.interpolate(inplace=True)
pt_refined_Z.plot(figsize=(10,5))
```
