```python
"""
    브로드캐스팅 소개
"""

a = np.array([0,1,2])
print(a + 5, '\n')

# 브로드캐스팅: 1D + 2D 배열
a = np.ones((3,3))
print(a, '\n')

b = np.arange(3)
print(b, '\n')

print(a + b)
```

```python
"""
    두 배열 모두 브로드캐스팅한 경우
""" 
a = np.arange(3)

# 차원 추가
a = a[:, np.newaxis]   # 1차원 행렬을 2차원으로 바꿈
print(a, '\n')

b = np.arange(3)
print(b, '\n')

print(a + b)
```

```python
"""
    브로드캐스팅 규칙
"""

# ex) 1D array + 2D array
a = np.ones((2,3))
b = np.arange(3)
print(a, '\n', b,'\n')

print(a + b, '\n')

# 두 배열의 형상
print(a.shape == (2,3))   # 1. 규칙 1에 따라 b.shape==(1,3)
print(b.shape == (3, ))   # 2. 규칙 2에 따라 b.shape==(2,3)
```

```python
# ex) broadcast both arrays
a = np.arange(3)
print(a, '\n')
a = a.reshape((3,1))
b = np.arange(3)
print(a, '\n', b, '\n')

print(a + b, '\n')

print(a.shape, b.shape)
```

```python
# ex) 브로드캐스팅이 불가능한 경우
a = np.ones((3,2))
print(a, '\n')
b = np.arange(3)
print(b, '\n')

print(a.shape, b.shape)

print(a + b)
```

```python
# 다차원 배열의 비교 연산
np.random.seed(0)
x = np.random.randint(10, size=(4,6))
print(x, '\n')

print(x > 5)
```

```python
# Boolean 배열 연산
print(np.count_nonzero(x < 6))
print(np.sum(x < 6))

# np.any()
print(np.any(x > 9))
print(np.any(x >= 9))

# np.all()
print(np.all(x > 0))
print(np.all(x >= 0))
```

```python
# Boolean 배열을 이용한 데이터 선택 - 마스킹 연산 Masking Operations
print(x, '\n')
print(x[x > 5])
```

```python
"""
    팬시 인덱싱(fancy indexing)
"""

np.random.seed(1)
x = np.random.randint(10, size=10)
print(x)

# 기본 문법
a = x[2]
b = x[4]
c = x[6]
print(a, b, c)

# 팬시 인덱싱
ind = [2, 4, 6]
print(x[ind])
```

```python
# 팬시 인덱싱: 다차원 배열
np.random.seed(2)
x = np.random.randint(10, size=(3, 5))
print(x)

row = [0, 1, 2]
col = [2, 1, 4]

print(x[row, col])   # [0,2] [1,1] [2,4] 인덱싱
```

```python
# 팬시 인덱싱: 다차원 배열 + 브로드캐스팅 = 다차원 배열
print(x, '\n')

# 브로드캐스팅은 NumPy 배열에서 동작
row = np.array([0, 1, 2])
col = np.array([2, 1, 4])

print(row, '\n', row[:, np.newaxis], '\n')
print(x[row[:, np.newaxis], col])
```

```python
# 결합 인덱싱

# 3X4 2차원 배열(1, 12) 생성
x = np.arange(12) + 1
x = x.reshape((3,4))
print(x)

ind = [2, 0, 1]

print(x[2, ind])
print(x[1:, ind])
```

```python
"""
    예제: 팬시 인덱싱을 이용한 임의의 점 집합 선택하기
"""

# 시각화 설정
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()  # set plot styles
%config InlineBackend.figure_format = 'svg'
```

```python
mean = [0, 0]     # 평균
cov = [[1, 2],    # 공분산
      [2, 5]]

# np.random.seed(1) : 컴퓨터가 계속 seed=1로 난수 생성. 계속 바꾸면 헷갈리고 비효율적
# RandomState: 시드를 별도 객체에서 관리 - 여러개 seed 사용 시 편리
rand = np.random.RandomState(1)

# multivariate_normal(): 정규분포 다변량 데이터 생성
x = rand.multivariate_normal(mean, cov, 100)   # 평균과 공분산에 맞춰서 100개

print(x[:10])
print(x.shape)
```

```python
# 산포도 출력: scatter(x, y)
plt.scatter(x[:, 0], x[:, 1])
```

```python
# 20개의 점을 무작위 추출(random sampling) -> 인덱스 값을 리턴
ind = np.random.choice(x.shape[0], 20, replace=False)   # x.shape = (100, 2)  # replace=False : 중복 없애기
print(ind)
```

```python
# 선택된 점 집합 출력
selected = x[ind]

print(selected)
print(selected.shape)
```

```python
# 선택된 점 집합 시각화
plt.scatter(x[:, 0], x[:, 1], alpha=0.3)    # 원본 점 집합
plt.scatter(selected[:, 0], selected[:, 1], # 선택된 점 집합
            facecolor='red', s=50);
```
